```python
# Import Dependencies

import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import key_list
```

```python
# Read CSV file 
data_csv = ("data/2015.csv")
data1 = pd.read_csv(data_csv)
d1_csv = ("data/2016.csv")
data2 = pd.read_csv(d1_csv)
d_csv = ("data/2017.csv")
data3 = pd.read_csv(d_csv)
```

```python
data3.head()
```

```python
merge_df = pd.merge(data1, data2, on="Country", how="left")
merge_df1= pd.merge(merge_df, data3, on="Country", how="left")
```

```python
merge_df1.head()
```

```python
# Slice columns, rename and reset index

cleaned_df = merge_df1[['Country','Region_x','Happiness Rank_x','Happiness Score_x','Freedom_x','Family_x','Standard Error','Trust (Government Corruption)_x','Health (Life Expectancy)_x', 'Economy (GDP per Capita)_x', 'Generosity', 'Dystopia.Residual', 'Whisker.high', 'Whisker.low' ]]
happiness_df1 = cleaned_df.rename(columns={'Region_x': 'Region', 'Happiness Rank_x': 'Happiness_Rank','Happiness Score_x': 'Happiness_Score','Standard Error': 'Standard_Error', 'Freedom_x' : 'Freedom', 'Family_x': 'Family', 'Trust (Government Corruption)_x': 'Trust(Government_Corruption)', 'Health (Life Expectancy)_x': 'Health(Life_Expectancy)','Economy (GDP per Capita)_x': 'Economy(GDP_per_capita)','Dystopia.Residual': 'Dystopia_Residual', 'Whisker.high': 'Whisker_high', 'Whisker.low': 'Whisker_low'})
happiness_df = happiness_df1.sort_values('Country', ascending=True)
happiness_df.reset_index(inplace=True)
happiness_df.drop(columns={'index'}, inplace=True)
happiness_df.head()
```

```python
# Add columns for lat, lng

happiness_df["Lat"] = ""
happiness_df["Lng"] = ""
happiness_df.head()
```

```python
g = 0
params = {"key": key_list[0]}
i = 0
# Loop through the cities_pd and run a lat/long search for each city
for index, row in happiness_df.iterrows():
    try:
        #i += 1
        base_url = "https://maps.googleapis.com/maps/api/geocode/json"

        country = row['Country']
        
        # update address key value
        params['address'] = f"{country}"

        # make request, print url
        country_lat_lng = requests.get(base_url, params=params)
        print(country_lat_lng.url)

        # convert to json
        country_lat_lng = country_lat_lng.json()

        happiness_df.set_value(
            index, "Lat", country_lat_lng["results"][0]["geometry"]["location"]["lat"])
        happiness_df.set_value(
            index, "Lng", country_lat_lng["results"][0]["geometry"]["location"]["lng"])
        #print(i)
    except IndexError:
        g += 1
        params = {"key": key_list[g]}
        continue
        
```
